```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## An image gradient is the directional change in the intensity or color in an image- used for edge detection

## 1.Laplacian derivative-lapace derivative is used
## 2.sobelX and sobelY

```python
img=cv2.imread('lenna.png', 0)

lap=cv2.Laplacian(img, cv2.CV_64F, ksize=3) # arg-source, datatype(64 float type that support negative numbers)
lap=np.uint8(np.absolute(lap)) # take abs image and chnage datatype to int


# Change is considered in x direction
sobelX=cv2.Sobel(img, cv2.CV_64F,1,0)  #arg- source, datatype, dx(sobelX), dy(sobelY)
sobelX=np.uint8(np.absolute(sobelX))

#Change is considered in y direction
sobelY=cv2.Sobel(img, cv2.CV_64F,0,1)
sobelY=np.uint8(np.absolute(sobelY))

# to combine the results of the sobelX and sobelY
sobelCombined=cv2.bitwise_or(sobelX, sobelY)

titles=['image','laplacian','sobelX', 'sobelY','sobelCombined']
images=[img, lap,sobelX,sobelY,sobelCombined]

for i in range(len(titles)):
    plt.subplot(2,3,i+1),plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

    
plt.show()
```

### Canny edge detector is an edge detection operator that uses a multi-stage algorithm to detect a wide range of edges in images

## 5 steps- 
###    1.Noise reduction by applying gaussian filter
###    2.find Gradients
###    3.Non maximum suppresion
###    4.double threshold
###    5.edge tracking by hysteresis


```python
img=cv2.imread('lenna.png', 1)
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# arg-source, first threshold value, second threshold value, 
canny=cv2.Canny(img, 100,200) 


titles=['image','canny']
images=[img, canny]

for i in range(len(titles)):
    plt.subplot(1,2,i+1),plt.imshow(images[i],'gray')
    plt.xticks([]), plt.yticks([])
    plt.title(titles[i])
    
plt.show()
```

## we can see that canny edge detection gives us better result - i.e it contains less noise
