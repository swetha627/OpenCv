## Smoothing/blurring -  used for removing noise in the images

## Linear filters are fast-
    1.Homogenous
    2.gaussian
    3.bilateral

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Homogenous filter - each ouput pixel is the mean of its kernel neighbors
    kernel for homogenous filter = [1 / k_width*k_height](kernel)

```python
img=cv2.imread('water.jfif')
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel=np.ones((2,2), np.float32)/4   # 4-(2*2) should be divided because of the kernel formula given above
output=cv2.filter2D(img, -1, kernel)   # arg-image source, desired depth of output image, kernel 


titles=['image', '2D Convolution']
images=[img, output]


for i in range(2):
    plt.subplot(1,2,i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()
```

## As in one dimension signals, images can also be filtered using high pass and low pass filters

## low pass filter helps in smoothing or blurring the images
## high pass filter helps in finding edges in the image

```python
# BLUR or Averaging method 
blur=cv2.blur(img, (2,2))
plt.imshow(blur, 'gray')
plt.show()
```

## Gaussian filter is nothing but using different-weight-kernel in both x and y direction

## pixels located in the center of kernel have higher weight and weight decreases as we move toward ends the the kernel

## gaussian blur specifically designed to remove high frequency noise from the images 

```python

gBlur=cv2.GaussianBlur(img, (3,3),0) # arg -image, kernel, sigma value
plt.imshow(gBlur, 'gray')
plt.show()
```

## Median fliter- This is helpfull in removing "salt and pepper noise".
In this technique each pixel is replaced with median of the neighboring pixels

```python
img1=cv2.imread('water.jfif')
median=cv2.medianBlur(img1,5) #arg-source, kernel size- must odd number except 1
images=[img1, median]
titles=['before fliter', 'after filter']
for i in range(2):
    plt.subplot(1,2, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.show()
# for salt and pepper noise median filter is the best 
```

## Bilateral filter- remove noise from the images without smoothing edges - i.e edges remain sharper even if the image is blurred

```python
img2=cv2.imread('lenna.png')
img2=cv2.cvtColor(img2 ,cv2.COLOR_BGR2RGB)
# arg-source, diameter of the each pixel from the neighbor hood, sigma color, sigma sapce 
bilateralFilter=cv2.bilateralFilter(img2,9, 75, 75)

plt.imshow(bilateralFilter, 'gray')
plt.show()
```
