# matplotlib is a python 2D ploting library 

```python
import cv2
from matplotlib import pyplot as plt
```

```python
# openCv reads image in BGR format
img=cv2.imread('lenna.png')
cv2.imshow('image', img) 
#   img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# matplotlib reads image in RGB format 
plt.imshow(img)
# plt.xticks([]), plt.yticks([]) # to hide x and y axis
plt.show()


cv2.waitKey(0)
cv2.destroyAllWindows()
```

# subplots 

```python
img=cv2.imread('gradient.jfif', 0)
_,th1=cv2.threshold(img, 50,255, cv2.THRESH_BINARY)
_,th2=cv2.threshold(img, 200,255, cv2.THRESH_BINARY_INV)
_,th3=cv2.threshold(img, 127,255, cv2.THRESH_TRUNC)
_,th4=cv2.threshold(img, 127,255, cv2.THRESH_TOZERO)
_,th5=cv2.threshold(img, 50,255, cv2.THRESH_TOZERO_INV)

titles=['Original', 'BINARY', 'BINARY_INV', 'TRUNC', 'TOZERO', 'TOZERO_INV']
images=[img, th1, th2, th3, th4, th5]

for i in range(6):
    plt.subplot(2,3,i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()
```
