```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# Morphological Transformation are simple Operations based on image shapes 

```python
# morphological operations are performed on BINARY images 
```

# DILATION-  pixels are added around the boundaries of the original image

```python
img=cv2.imread('smarties.jfif',0)
_,mask=cv2.threshold(img, 220, 255, cv2.THRESH_BINARY_INV)

# kernal is some shape which we want to apply on the image for example a rectangle
# if atleast one pixel in kernel range of image is 1, then whole space under that kernel will be one  
# if we increase size of the kernel then objects in the image starts to merge

kernal=np.ones((2,2), np.uint8) 
dilation=cv2.dilate(mask, kernal, iterations=2)

titles=['image', 'mask', 'dilation']
images=[img, mask, dilation]

for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

    
plt.show()
```

# EROSION- pixels around the boundaries of original image are removed

```python
img=cv2.imread('smarties.jfif',0)
_,mask=cv2.threshold(img, 220, 255, cv2.THRESH_BINARY_INV)

kernel=np.ones((2,2), np.uint8)
erosion=cv2.erode(mask, kernel,iterations=2)


titles=['image', 'mask', 'erosion']
images=[img, mask, erosion]

for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

    
plt.show()
```

## OPENING - Erosion followed by dilation

```python

kernel=np.ones((5,5), np.uint8)
opening=cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

titles=['image', 'mask', 'opening']
images=[img, mask, opening]


for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

    
plt.show()
```

## CLOSING- Dilation followed by Erosion

```python

kernel=np.ones((5,5), np.uint8)
closing=cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

titles=['image', 'mask', 'closing']
images=[img, mask, closing]


for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

    
plt.show()
```

### Morphological gradient - difference between dilation and erosion of an image



### two types of TopHat- BlackHat and WhiteHat 

### WhiteHat - difference between image and opening of the image


### BlackHat - difference between image and closing of the image

```python
mor_gd=cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
topHat=cv2.morphologyEx(img, cv2.MORPH_TOPHAT,kernel)
blackHat=cv2.morphologyEx(img, cv2.MORPH_BLACKHAT,kernel)

images=[mor_gd, topHat, blackHat]
titles=['mor_gd', 'topHat', 'blackHat']

for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
    
```
